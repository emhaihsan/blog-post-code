```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import CharacterTextSplitter
```

```python
loaders = [
    PyPDFLoader("/kaggle/input/indonesia-travel-guide-documents/adventuredestination.pdf"),
    PyPDFLoader("/kaggle/input/indonesia-travel-guide-documents/romanticdestination.pdf"),
    PyPDFLoader("/kaggle/input/indonesia-travel-guide-documents/prioritydestination.pdf"),
    PyPDFLoader("/kaggle/input/indonesia-travel-guide-documents/familytravel.pdf"),
    PyPDFLoader("/kaggle/input/indonesia-travel-guide-documents/thrillingadventures.pdf"),
]
```

```python
pages = []
for loader in loaders:
    pages.extend(loader.load())
```

```python
len(pages)
```

```python
text_splitter = CharacterTextSplitter(
    separator="\n",
    chunk_size=500,
    chunk_overlap=100,
)
```

```python
docs = text_splitter.split_documents(pages)
```
